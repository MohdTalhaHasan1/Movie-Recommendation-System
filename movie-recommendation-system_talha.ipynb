# Movie Recommendation

This project aimed to build a movie recommendation system that suggests movies you might like based on what similar users enjoyed. It used a technique called collaborative filtering to analyze user ratings and find patterns. The project evaluated the system's accuracy by measuring how well it predicted actual ratings, but also acknowledged the need for a more comprehensive evaluation in the future. 

## Types of recommendation system
 - Popularity Based 

It keeps a track of view counts for each movie/video and then lists movies based on views in descending order.

 - Content Based
 
This type of recommendation systems, takes in a movie that a user currently likes as input. Then it analyzes the contents of the movie to find out other movies which have similar content. Then it ranks similar movies according to their similarity scores and recommends the most relevant movies to the user. 

 - Collaborative filtering

In other words, the recommendations get filtered based on the collaboration between similar userâ€™s preferences.

In this Project, we are going to implement both content based  and Collaborative filtering recommendation system.

## Content Based Recommendation System

```python
cv=TfidfVectorizer()
tfidf_matrix=cv.fit_transform(movies['genres'])
```

```python
movie_user = df.pivot_table(index='userId',columns='title',values='rating')
movie_user.head()
```

Suppose a user wants to watch a movie similar to Toy Story (1995) then we can recommend the user by calculating the cosine similarity between Toy Story and other movies. So we have to first find the cosine similarity betw

```python

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
indices=pd.Series(movies.index,index=movies['title'])
titles=movies['title']
def recommendations(title):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:21]
    movie_indices = [i[0] for i in sim_scores]
    return titles.iloc[movie_indices]
```

```python
recommendations('Toy Story (1995)')
```

## Collabrative Filtering Recommendation System

```python
df.columns
```

```python
user_item_matrix = csr_matrix((df['rating'], (df['userId'], df['movieId'])))
```

```python
user_similarity = cosine_similarity(user_item_matrix)
```

```python
knn = NearestNeighbors(n_neighbors=10, metric='cosine')  # Choose K (e.g., 10, 20)
knn.fit(user_similarity)
```

```python
def predict_rating(user_id, movie_id):
  neighbors = knn.kneighbors(user_similarity[user_id].reshape(1, -1), return_distance=False)[0]
  ratings = user_item_matrix[neighbors, movie_id].toarray().flatten()
  similarities = user_similarity[user_id, neighbors].flatten()
  weighted_average = np.average(ratings, weights=similarities)
  return weighted_average
```

```python
def recommend_movies(user_id, N=10):
  unrated_movies = user_item_matrix[user_id].toarray()[0].nonzero()[0]
  predicted_ratings = [predict_rating(user_id, movie_id) for movie_id in unrated_movies]
  movie_recommendations = np.argsort(predicted_ratings)[::-1][:N]
  return movies[movies['movieId'].isin(movie_recommendations)]
```



**A movie recommendation system based on collaborative filtering.**

* It leverages a user-item matrix to store user ratings for different movies.
* The `predict_rating` function takes a user ID and movie ID as input. It identifies the user's nearest neighbors (most similar users based on ratings) and calculates a weighted average of their ratings for the unrated movie. This weighted average represents the predicted rating the user might give to the movie.
* The `recommend_movies` function utilizes the `predict_rating` function to recommend movies to a user. It first identifies movies the user hasn't rated. Then, it predicts ratings for those unrated movies and recommends the top N movies (by default, 10) with the highest predicted ratings to the user.

**In essence, this code recommends movies to users by considering the preferences of similar users (those who have rated movies similarly in the past).**


```python
recommend_movies(3)
```

## Model Evaluation

```python
from sklearn.model_selection import train_test_split
# Split data into training and testing sets (e.g., 80% training, 20% testing)
train_data, test_data = train_test_split(df, test_size=0.2, random_state=42)
train_user_ids = train_data['userId'].unique()
test_user_ids = test_data['userId'].unique()
```

```python
#Define functions to calculate evaluation metrics
def rmse(predictions, targets):
  return sqrt(mean_squared_error(predictions, targets))

def mae(predictions, targets):
  return mean_absolute_error(predictions, targets)




```

```python
# Evaluate the model on the testing set
all_predictions = []
all_targets = []
for user_id in test_user_ids:
  unrated_movies = test_data[test_data['userId'] == user_id]['movieId'].tolist()
  predictions = [predict_rating(user_id, movie_id) for movie_id in unrated_movies]
  targets = test_data[test_data['userId'] == user_id]['rating'].tolist()
  all_predictions.extend(predictions)
  all_targets.extend(targets)
```

```python

# Calculate RMSE, MAE
rmse_score = rmse(all_predictions, all_targets)
mae_score = mae(all_predictions, all_targets)


print("RMSE:", rmse_score)
print("MAE:", mae_score)
```

RMSE (Root Mean Squared Error): It measures the average magnitude of the difference between predicted ratings and actual ratings. In your case, 1.877 indicates that, on average, the predicted ratings differ from the actual ratings by about 1.88 on a scale of 1.5 to 4 (assuming your rating scale). While lower RMSE is better (ideally closer to 0), interpreting the value depends on your rating scale. A difference of 1.88 on a 5-point scale might be less concerning than the same difference on a 10-point scale.
